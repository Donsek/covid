```python
df_europe = df[df['Регион ВОЗ'] == 'Европейский регион']
df_europe.head()
```

```python
df['%_подтверждено_за_неделю'] = (
    (df['Подтверждено случаев'] - df['Подтверждено за прошлую неделю'])
    / df['Подтверждено за прошлую неделю']
    * 100
)
```

```python
df['%_смертей_за_неделю'] = (
    df['Новые смерти'] / (df['Смертей'] - df['Новые смерти']).replace(0, pd.NA) * 100
)
```

```python
df['%_выздоровлений_за_неделю'] = (
    df['Новые выздоровления'] / (df['Выздоровело'] - df['Новые выздоровления']).replace(0, pd.NA) * 100
)
```

```python
df['Смертей_на_100_случаев'] = (
    df['Смертей'] / df['Подтверждено случаев'].replace(0, pd.NA) * 100
)
```

```python
df['Выздоровлений_на_100_случаев'] = (
    df['Выздоровело'] / df['Подтверждено случаев'].replace(0, pd.NA) * 100
)
```

```python
df['Смертей_на_100_выздоровлений'] = (
    df['Смертей'] / df['Выздоровело'].replace(0, pd.NA) * 100
)
```

```python
cols = [
    'Страна/Регион',
    'Регион ВОЗ',
    '%_подтверждено_за_неделю',
    'Смертей_на_100_случаев',
    'Выздоровлений_на_100_случаев',
    'Смертей_на_100_выздоровлений',
]
df[cols].head(10)
```

```python
pivot_who = pd.pivot_table(
    df,
    index='Регион ВОЗ',
    values=['Подтверждено случаев', 'Смертей', 'Выздоровело', 'Активные случаи'],
    aggfunc='sum'
).sort_values(by='Подтверждено случаев', ascending=False)

pivot_who
```

```python
df['Уровень смертности'] = (
    df['Смертей'] / df['Подтверждено случаев'].replace(0, pd.NA) * 100
)
```

```python
pivot_countries = pd.pivot_table(
    df,
    index='Страна/Регион',
    values='Уровень смертности',
    aggfunc='mean'
).sort_values(by='Уровень смертности', ascending=False)

pivot_countries.head(10)
```

```python
import matplotlib.pyplot as plt

# Берём топ-15 стран, чтобы график был читаем
top_new_cases = (
    df[['Страна/Регион', 'Новые случаи']]
    .sort_values('Новые случаи', ascending=False)
    .head(15)
)

plt.figure(figsize=(12, 6))
plt.bar(top_new_cases['Страна/Регион'], top_new_cases['Новые случаи'])
plt.xticks(rotation=45, ha='right')
plt.title('Новые случаи COVID-19 по странам')
plt.xlabel('Страна')
plt.ylabel('Новые случаи')
plt.tight_layout()
plt.show()
```

```python
pivot_who = (
    df
    .groupby('Регион ВОЗ', as_index=False)
    .agg(Подтверждено=('Подтверждено случаев', 'sum'))
)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(
    pivot_who['Подтверждено'],
    labels=pivot_who['Регион ВОЗ'],
    autopct='%1.1f%%',
    startangle=140
)
plt.title('Распределение подтверждённых случаев COVID-19 по регионам ВОЗ')
plt.tight_layout()
plt.show()
```
